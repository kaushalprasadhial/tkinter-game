```python
from tkinter import *
import tkinter.messagebox as tkmbox
import time
import random

root = Tk()
root.title('bouce back')
root.resizable(0,0)
root.wm_attributes('-topmost',1)
hight = 400
width = 400
canvas = Canvas(root, width=width, height=hight, bg='gray', bd=0 , highlightthickness=0)
canvas.pack()
root.update()

class Ball:
    def __init__(self,canvas,paddel,color):
        self.canvas = canvas
        self.paddel = paddel
        self.oval = canvas.create_oval(10,10,25,25, fill = color)
        self.canvas.move(self.oval,200,200)
        start = [-3,-2,-1,0,1,2,3]
        random.shuffle(start)
        self.x = start[0]
        self.y = -3
        self.canvas_height = self.canvas.winfo_height()
        self.canvas_width = self.canvas.winfo_width()
        print('canvas height', self.canvas_height)
        print('canvas width', self.canvas_width)
        
    def hit_paddel(self, pos):
        paddel_pos = self.canvas.coords(self.paddel.paddel)
        if pos[2]>=paddel_pos[0] and pos[0]<=paddel_pos[2]:
            if pos[3]>=paddel_pos[1] and pos[1]<=paddel_pos[3]:
                return True
            return False
        return False
       
    def move(self):
        self.canvas.move(self.oval,self.x,self.y)
        pos = self.canvas.coords(self.oval)
        print(pos)
        print('x : ', self.x)
        print('y : ', self.y)
        if pos[1]<=0:
            self.y = 2
        if pos[3]>=self.canvas_height:
            self.y =-2
        if pos[0]<=0:
            self.x = 2
        if pos[2]>=self.canvas_width:
            self.x = -2
        if self.hit_paddel(pos):
            self.y=-3

class Paddel:
    def __init__(self, canvas, color):
        self.canvas = canvas
        self.paddel = canvas.create_rectangle(0,0,100,10, fill=color)
        self.canvas.move(self.paddel, 200,300)
        self.x = 0
        self.canvas_width = self.canvas.winfo_width()
        self.canvas.bind_all('<KeyPress-Left>',self.turn_left)
        self.canvas.bind_all('<KeyPress-Right>',self.turn_right)
        
    def move(self):
        self.canvas.move(self.paddel, self.x, 0)
        pos = self.canvas.coords(self.paddel)
        if pos[0]<=0:
            self.x = 0
        if pos[2]>=self.canvas_width:
            self.x=0
            
    def turn_left(self,evt):
        self.x = -2
        
    def turn_right(self,evt):
        self.x = 2
    
    
paddel = Paddel(canvas, 'red')
ball = Ball(canvas, paddel, 'black')

while 1:
    ball.move()
    paddel.move()
    root.update_idletasks()
    root.update()
    time.sleep(.04)
    
root.mainloop()
"""try:
    while 1:
        ball.move()
        root.update_idletasks()
        root.update()
        time.sleep(.01)

    root.mainloop()
except:
    tkmbox.showinfo('something went wrong')
    root.destroy()"""
```
