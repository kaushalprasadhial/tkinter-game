```python
from tkinter import *
import time
import random

root = Tk()
root.title('ping-pong')
root.resizable(0,0)
root.wm_attributes('-topmost',1)


height = 500
width = 700

canvas = Canvas(root, height = height, width = width)
canvas.config(bg='black')
canvas.pack()

frame =Frame(root, height = 100, width = canvas.winfo_width(), bg = '#3399ff')
frame.pack(fill = X)

label = Label(frame, text = 'Player One: 0')
label.pack(side=LEFT, fill = X, expand = True)

label2 = Label(frame,text = '5 miss opposite player wins')
label2.pack(side=LEFT, fill = BOTH, expand = True)

label3 = Label(frame, text = 'Player Two: 0')
label3.pack(side=LEFT, fill = X, expand = True)

canvas.create_line(width/2,0,width/2,height,fill='white')
root.update()


#paddel class
class Paddel_left:
    def __init__(self, canvas, color):
        self.canvas = canvas
        self.canvas_width = self.canvas.winfo_width()
        self.canvas_height = self.canvas.winfo_height()
        self.paddel = canvas.create_rectangle(0,0,20,80, fill=color)
        self.canvas.move(self.paddel, 0,0)
        self.y = 0
        self.pos = self.canvas.coords(self.paddel)
        self.canvas.bind_all('a',self.turn_up)
        self.canvas.bind_all('z',self.turn_down)
        
    def move(self):
        self.canvas.move(self.paddel, 0, self.y)
        self.pos = self.canvas.coords(self.paddel)
        if self.pos[1]<=0:
            self.y = 0
        if self.pos[3]>=self.canvas_height:
            self.y=0
            
    def turn_up(self,evt):
        self.y = -7
        if self.pos[1]<=0:
            self.y =0
        
    def turn_down(self,evt):
        self.y = 7
        if self.pos[3]>=self.canvas_height:
            self.y =0
        
#right paddel class
class Paddel_right:
    def __init__(self, canvas, color):
        self.canvas = canvas
        self.canvas_width = self.canvas.winfo_width()
        self.canvas_height = self.canvas.winfo_height()
        self.paddel = canvas.create_rectangle(0,0,20,80, fill=color)
        self.canvas.move(self.paddel, self.canvas_width-20,0)
        self.y = 0
        self.pos = self.canvas.coords(self.paddel)
        self.canvas.bind_all('<KeyPress-Up>',self.turn_up)
        self.canvas.bind_all('<KeyPress-Down>',self.turn_down)
  
    def move(self):
        self.canvas.move(self.paddel, 0, self.y)
        self.pos = self.canvas.coords(self.paddel)
        if self.pos[1]<=0:
            self.y = 0
        if self.pos[3]>=self.canvas_height:
            self.y=0
            
    def turn_up(self,evt):
        self.y = -7
        if self.pos[1]<=0:
            self.y=0
        
    def turn_down(self,evt):
        self.y = 7
        if self.pos[3]>=self.canvas_height:
            self.y =0
        
#ball class
class Ball:
    def __init__(self,canvas,paddel_left,paddel_right,color):
        self.left_score = 0
        self.right_score = 0
        self.anybody_won = False
        self.canvas = canvas
        self.paddel_left = paddel_left
        self.paddel_right = paddel_right
        self.oval = canvas.create_oval(10,10,25,25, fill = color)
        self.canvas.move(self.oval,width/2,height/2)
        start = [-3,3]
        random.shuffle(start)
        self.x = start[0]
        self.y = -3
        self.canvas_height = self.canvas.winfo_height()
        self.canvas_width = self.canvas.winfo_width()
        
    def won(self, paddel):
        self.stop_everything()
        if(paddel==self.paddel_left):
            label2['text']=( 'Player Two Won The Game! Game Over')
        else:
            label2['text']=( 'Player One Won The Game! Game Over')
        self.anybody_won = True
            
    def stop_everything(self):
        self.x = 0 
        self.y = 0
        self.paddel_left.canvas.bind_all('a',self.do_nothing)
        self.paddel_left.canvas.bind_all('z',self.do_nothing)
        self.paddel_left.y = 0
        self.paddel_right.canvas.bind_all('<KeyPress-Up>',self.do_nothing)
        self.paddel_right.canvas.bind_all('<KeyPress-Down>',self.do_nothing)
        self.paddel_right.y = 0
        
    def do_nothing(self, evnt):
        pass
        
    def hit_paddel_left(self, pos):
        paddel_pos = self.canvas.coords(self.paddel_left.paddel)
        if pos[2]>=paddel_pos[0] and pos[0]<=paddel_pos[2]:
            if pos[3]>=paddel_pos[1] and pos[1]<=paddel_pos[3]:
                return True
            return False
        return False

    def hit_paddel_right(self, pos):
        paddel_pos = self.canvas.coords(self.paddel_right.paddel)
        if pos[2]>=paddel_pos[0] and pos[0]<=paddel_pos[2]:
            if pos[3]>=paddel_pos[1] and pos[1]<=paddel_pos[3]:
                return True
            return False
        return False
       
    def move(self):
        if not self.anybody_won:
            self.canvas.move(self.oval,self.x,self.y)
            pos = self.canvas.coords(self.oval)
            
            if pos[1]<=0:
                self.y = 5
            if pos[3]>=self.canvas_height:
                self.y =-5
            if pos[0]<=0:
                self.x = 5
                self.left_score += 1
                label['text']=( 'Player One: '+ str(self.left_score))
                if self.left_score >= 5:
                    self.won(self.paddel_left)
            if pos[2]>=self.canvas_width:
                self.x = -5
                self.right_score += 1
                label3['text']=( 'Player Two: '+ str(self.right_score))
                if self.right_score>=5:
                    self.won(self.paddel_right)
            if self.hit_paddel_left(pos):
                self.x = 5
            if self.hit_paddel_right(pos):
                self.x = -5
    

            
paddel_left = Paddel_left(canvas, 'red')
paddel_right = Paddel_right(canvas, 'red')
ball = Ball(canvas, paddel_left, paddel_right, 'blue')

while 1:
    root.update()
    ball.move()
    paddel_left.move()
    paddel_right.move()
    root.update_idletasks()
    
    time.sleep(.04)

root.mainloop()
```
